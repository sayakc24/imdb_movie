## 1.  Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
# Replace extremely low value of budget and revenue with the median value
imdb_new_df.loc[imdb_new_df['budget'] < 1000, 'budget'] = imdb_new_df.loc[imdb_new_df['budget'] >= 1000, 'budget'].median()
imdb_new_df.loc[imdb_new_df['revenue'] < 1000, 'revenue'] = imdb_new_df.loc[imdb_new_df['revenue'] >= 1000, 'revenue'].median()
```

### Profit can be calculated as the difference between budget and revenue

```python
imdb_new_df['profit'] = imdb_new_df['revenue'] - imdb_new_df['budget']
```

# 2. This data has information about movie made in different languages. Which language has the highest averga ROI?

```python
# Calculate the ROI
imdb_new_df['roi'] = 100 * imdb_new_df['profit'] / imdb_new_df['budget']
```

```python
imdb_new_df.head()
```

### We will groupby on the original language and calculate the mean of the ROI

```python
# Creating grouped dataframe
roi_languages_df = imdb_new_df.groupby('original_language')['roi'].mean().reset_index()
```

```python
roi_languages_df.head()
```

```python
max_roi_lan = roi_languages_df[roi_languages_df['roi'] == roi_languages_df['roi'].max()]
```

```python
print('The language with the highest average ROI is :', max_roi_lan['original_language'].iloc[0])
```

# 4. Make a table of all producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average ROI. 

```python
# Create a dataframe in which we have no null observation for the column crew

imdb_df_no_na_crew = imdb_df[~imdb_df['crew'].isna()]
```

```python
imdb_df_no_na_crew.iloc[0],['crew']
```

```python
# Creating a function which returns the list of producers and directors for each movie

def get_producers_and_directors(row):
    try:
        # Initialize two lists
        producer_list = []
        director_list = []
        for elem in row['crew']:
            if elem['job'] == 'Producer':
                producer_list.append(elem['name'])
            if elem['job'] == 'Director':
                director_list.append(elem['name'])
                row['producers'] = producer_list
                row['directors'] = director_list
        return row
    except:
        return None
    
```

```python
imdb_new2_df = imdb_new_df.apply(get_producers_and_directors, axis = 1)
```
